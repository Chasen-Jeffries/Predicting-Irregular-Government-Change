```python
import pandas as pd
import numpy as np
```

```python
Aid = pd.read_csv('Data/Clean/International/Aid.csv')
Aid.head()
```

```python
Exports = pd.read_csv('Data/Clean/International/Exports.csv')
Exports.head()
```

```python
Imports = pd.read_csv('Data/Clean/International/Imports.csv')
Imports.head()
```

```python
FDI = pd.read_csv('Data/Clean/International/FDI.csv')
FDI.head()
```

```python
Mil_Exp = pd.read_csv('Data/Clean/International/ME.csv')
Mil_Exp.head()
```

```python
# Updating Year Column
Mil_Exp = Mil_Exp.rename(columns={Mil_Exp.columns[1]: 'Year'})
```

```python
# Storing all DataFrames in a list
dfs = [FDI, Imports, Exports, Aid, Mil_Exp]  # Add more DataFrames to this list

# Initialize the merged DataFrame with the first DataFrame in the list
main_df = dfs[0]

# Iteratively merge each DataFrame in the list
for df in dfs[1:]:
    main_df = pd.merge(main_df, df, on=['Country', 'Year'])
```

```python
main_df.head()
```

```python
# Exporting Combined df
# main_df.to_csv('Data/Clean/Combined/International-Data.csv', index = False)
```

```python
# Bringing in the DV
PI = pd.read_csv('Data/Clean/Political/PI.csv')
PI.head()
```

```python
final_df = pd.merge(main_df, PI, on=['Country', 'Year'])
```

```python
final_df.head()
```

```python
# Exporting Combined df
final_df.to_csv('Data/Clean/Combined/International-Data.csv', index = False)
```
