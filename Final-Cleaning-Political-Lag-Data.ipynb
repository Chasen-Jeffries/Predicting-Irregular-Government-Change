```python
import pandas as pd
import numpy as np
```

```python
Political_Data = pd.read_csv('Data/Clean/Combined/Political-Data.csv')
Political_Data.head()
```

```python
# Create DataFrame
df_01 = pd.DataFrame(Political_Data)

# Set both 'Year' and 'Country' as indices
df_01.set_index(['Year', 'Country'], inplace=True)

# Lag DV by 1 year
df_01['Gov'] = df_01['Gov'].shift(1)
df_01['BC'] = df_01['BC'].shift(1)
df_01['RPE'] = df_01['RPE'].shift(1)
df_01['PR'] = df_01['PR'].shift(1)
df_01['CL'] = df_01['CL'].shift(1)
df_01['Urban'] = df_01['Urban'].shift(1)
df_01['Fertility'] = df_01['Fertility'].shift(1)
df_01.head()
```

```python
# Reset the index and keep the 'Year' as a column
df_01 = df_01.reset_index()
```

```python
df_01.columns
```

```python
# Filter the DataFrame for years 2000 to 2020 without using the index
df_01 = df_01[(df_01['Year'] >= 2000) & (df_01['Year'] <= 2020)]
```

```python
df_01.head()
```

```python
df_01.head()
```

```python
# Drop 'Country' and 'Year' columns
# df_01 = df_01.drop(['Country', 'Year'], axis=1)
df_01.head()
```

```python
# DF Lag 1 Year
df_01.to_csv('Data/Clean/Final/Political-01.csv', index = False)
```
