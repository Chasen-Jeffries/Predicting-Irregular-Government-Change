```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/Original/International/Military-Expenditures.csv',encoding='latin1')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Iterate through all columns except the first one
for col in df.columns[1:]:
    # Check if the column is of string type
    if df[col].dtype == 'object':
        # Remove all non-numeric characters except for decimals
        df[col] = df[col].str.replace(r'[^\d.]', '', regex=True)
```

```python
df.head()
```

```python
# Iterate through the columns
for col in df.columns:
    # Check if the column is of string type
    if df[col].dtype == 'object':
        # Replace instances of three consecutive periods with an empty string
        df[col] = df[col].str.replace(r'\.{3}', '', regex=True)
```

```python
df.head()
```

```python
# Transform to long format
clean_df = pd.melt(df, id_vars='Country', var_name='year', value_name='ME')
```

```python
clean_df.head()
```

```python
# Replace empty strings with NaN across the DataFrame
clean_df.replace('', np.nan, inplace=True)
```

```python
clean_df.head()
```

```python
clean_df.to_csv('Data/Clean/International/ME.csv', index = False)
```
