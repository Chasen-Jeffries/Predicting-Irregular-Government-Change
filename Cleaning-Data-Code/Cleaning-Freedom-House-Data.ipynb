```python
import pandas as pd
import numpy as np
```

```python
# Now read the entire dataset with the modified column names
fh = pd.read_csv('Data/Original/Political/Freedom-House.csv')
```

```python
# Cleaning any extra spaces
fh.iloc[0] = fh.iloc[0].str.strip()
fh.head()
```

```python
df = fh
cl = fh
```

```python
df.drop(df.columns[df.isin(['CL', 'Status']).any()], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Drop the first row
df.drop(df.index[0], inplace=True)
```

```python
# Rename the first column to 'Country'
df.columns.values[0] = 'Country'
df.head()
```

```python
# Strip spaces from column names (not from the first row)
df.columns = df.columns.str.strip()
```

```python
# Assuming that the remaining columns are years and the values are some indicators
df = pd.melt(df, id_vars=['Country'], var_name='Year', value_name='PR')
```

```python
df.head()
```

```python
PR = df
PR.head()
```

```python
# Creating CL Datset
cl.drop(cl.columns[cl.isin(['PR', 'Status']).any()], axis=1, inplace=True)
```

```python
# Drop the first row
cl.drop(cl.index[0], inplace=True)
```

```python
# Rename the first column to 'Country'
cl.columns.values[0] = 'Country'
cl.head()
```

```python
# Strip spaces from column names (not from the first row)
cl.columns = cl.columns.str.strip()
```

```python
# Assuming that the remaining columns are years and the values are some indicators
cl = pd.melt(cl, id_vars=['Country'], var_name='Year', value_name='CL')
```

```python
cl.head()
```

```python
CL = cl
```

```python
CL.head()
```

```python
Main_Data = pd.merge(PR, CL, how='inner', on=['Country','Year'])
```

```python
Main_Data.head()
```

```python
# Convert the column to string type
Main_Data['Year'] = Main_Data['Year'].astype(str)
```

```python
# Replace characters that are not digits or hyphen
Main_Data['Year'] = Main_Data['Year'].str.replace(r'[^\d-]', '', regex=True)
```

```python
# Remove everything before and including the hyphen
Main_Data['Year'] = Main_Data['Year'].str.replace(r'^.*-', '', regex=True)
```

```python
# Convert back to numeric type if necessary (e.g., to integer)
Main_Data['Year'] = pd.to_numeric(Main_Data['Year'], errors='coerce')
```

```python
# Get unique values in the 'Year' column
Main_Data['Year'].unique()
```

```python
Main_Data.to_csv('Data/Clean/Political/FH.csv', index = False)
```
